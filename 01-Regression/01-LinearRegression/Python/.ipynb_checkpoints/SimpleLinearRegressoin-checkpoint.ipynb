# What is SLR?

The key idea is to consider a dependent variable Y and see how it is affected by one or more independent variables, typically denoted by x.

$ y = β_0 + β_1 x $

where:

- y = Depedent variable
- x = Independent variable
- $β_0$ = Intrecept
- $β_1$ = slope.
